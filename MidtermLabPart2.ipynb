```python
# PART II omplete the following questions in python programs by the functions in 
#Matplotlib

#7 Plot two lines on one figure with the legend for each

import matplotlib.pyplot as plt

# line 1 points
x1 = [10,20]
y1 = [20,40]
# plotting the line 1 points 
plt.plot(x1, y1, label = "line 1")
# line 2 points
x2 = [10,20]
y2 = [40,10]
# plotting the line 2 points 
plt.plot(x2, y2, label = "line 2")

plt.xlabel('x - axis')
plt.ylabel('y - axis')

# title
plt.title('Plot two lines on one figure with the legend for each ')
# show a legend 
plt.legend()
# Display a figure.
plt.show()

```

```python
#8 Draw a 2d scatter plot for samples (X1, X2, y) in different colors for y = 0 and y = 1 

import matplotlib.pyplot as plt
import random

x1=[]
x2=[]
y=[]


for i in range(0,10):
	x1.append(random.randint(1,30))
	x2.append(random.randint(1,30))
	y.append(random.randint(0,1)) 
 
plt.scatter(x1,x2,c= y) 
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('scatter plot')

plt.show()
```
